```python
import pandas as pd
```

```python
accepts = pd.read_csv('archive/chefmozaccepts.csv')
parking = pd.read_csv('archive/chefmozparking.csv')
cuisines = pd.read_csv('archive/chefmozcuisine.csv')
ratings = pd.read_csv('archive/rating_final.csv')
places = pd.read_csv('archive/geoplaces2.csv')
users = pd.read_csv('archive/userprofile.csv')
```

```python
#1
accepts = accepts.groupby('placeID')['Rpayment'].apply(list).to_frame()

places = places.merge(accepts, how='inner', on='placeID')

places.head()
```

```python
#2
parking = parking.groupby('placeID')['parking_lot'].apply(list).to_frame()
places = places.merge(parking, how='inner', on='placeID')
places.head()
```

```python
cuisines = cuisines.groupby('placeID')['Rcuisine'].apply(list).to_frame()
places = places.merge(cuisines, how='inner', on='placeID')
places.head()
```

```python
#4
ratings = pd.read_csv('archive/rating_final.csv')
cuisines = pd.read_csv('archive/chefmozcuisine.csv')

ratings = ratings.merge(cuisines, how='inner', on='placeID')
ratings = ratings.groupby('Rcuisine')[['rating','food_rating','service_rating']]

ratings.describe()
```

```python
userpayment = pd.read_csv('archive/userpayment.csv')
userprofile = pd.read_csv('archive/userprofile.csv')

userprofile = userprofile.merge(userpayment, how='inner', on='userID')
userprofile = userprofile[['drink_level', 'Upayment']]
userprofile[userprofile['drink_level'] == 'social drinker'].value_counts()
```

```python
usercuisine = pd.read_csv('archive/usercuisine.csv')
userprofile = pd.read_csv('archive/userprofile.csv')

userprofile = userprofile.merge(usercuisine, how='inner', on='userID')
userprofile = userprofile[['interest', 'Rcuisine']]
userprofile[userprofile['interest'] == 'eco-friendly'].value_counts()
```

```python
userpayment = pd.read_csv('archive/userpayment.csv')
userprofile = pd.read_csv('archive/userprofile.csv')

userprofile = userprofile.merge(userpayment, how='inner', on='userID')
userprofile = userprofile[['activity', 'Upayment']]
userprofile[userprofile['activity'] == 'student'].value_counts()
```
